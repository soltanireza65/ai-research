## Linear: y = 2x

```python
x = np.linspace(0, 10, 100)
y = 2 * x

plt.plot(x, y, label='y = 2x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear function')
plt.legend()
plt.show()
```

## Quadratic: y = x²

```python
x = np.linspace(-3, 3, 100)
y = x**2

plt.plot(x, y, label='y = x²')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic function')
plt.legend()
plt.show()
```

## Easy exercise: f(x) = 5x + 2 at x = 0, 1, -1

```python
def f(x):
    return 5 * x + 2

for x in [0, 1, -1]:
    print(f'f({x}) = {f(x)}')
```

## Composition: f(g(5)) where f(x)=2x, g(x)=x-3

```python
def f(x):
    return 2 * x

def g(x):
    return x - 3

print('f(g(5)) =', f(g(5)))
```
