```python
import numpy as np
import matplotlib.pyplot as plt

from app.utils.plot import setup_plot_style

setup_plot_style()


def numerical_derivative(f, x, h=0.001):
    return (f(x + h) - f(x - h)) / (2 * h)
```

## Derivative of y = 2x is constant 2

```python
def f(x):
    return 2 * x

x = np.linspace(0, 10, 100)
plt.plot(x, f(x), label='f(x)=2x')
plt.plot(x, np.full_like(x, 2), '--', label="f'(x)=2")
plt.legend()
plt.show()

print('numerical at 3:', numerical_derivative(f, 3.0))
```

## Power rule: derivative of x³

```python
def f(x):
    return x ** 3

x_pts = np.linspace(-2, 2, 100)
plt.plot(x_pts, f(x_pts), label='f(x)=x³')
plt.plot(x_pts, 3 * x_pts**2, '--', label="f'(x)=3x²")
plt.legend()
plt.show()

print('numerical at 2:', numerical_derivative(f, 2.0))
```
