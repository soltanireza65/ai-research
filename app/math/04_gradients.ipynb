## Gradient of f(x,y) = x² + y²

```python
def gradient_f(x, y):
    return np.array([2 * x, 2 * y])

for pt in [(0, 0), (1, 1), (2, -1)]:
    print(pt, '->', gradient_f(*pt))

x = np.linspace(-2, 2, 8)
y = np.linspace(-2, 2, 8)
X, Y = np.meshgrid(x, y)
U, V = 2 * X, 2 * Y

plt.quiver(X, Y, U, V)
plt.gca().set_aspect('equal')
plt.title('Gradient field of x² + y²')
plt.show()
```
