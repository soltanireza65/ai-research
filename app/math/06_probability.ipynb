```python
import numpy as np
from collections import Counter
```

## Coin and die

```python
p_heads = 1 / 2
p_six = 1 / 6
print('P(heads) =', p_heads)
print('P(die=6) =', p_six)

rolls = np.random.randint(1, 7, size=10_000)
counts = Counter(rolls)
print('Simulated P(6) ≈', counts[6] / len(rolls))
```
