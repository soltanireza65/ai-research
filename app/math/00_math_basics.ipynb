## Exponents

```python
x = 4
print('x**2 =', x ** 2)
print('x**0.5 =', x ** 0.5)
```

## Slope as rise over run

```python
rise, run = 6, 3
print('slope =', rise / run)
```

## Axes: plot y = 2x

```python
import matplotlib.pyplot as plt
from app.utils.plot import setup_plot_style
setup_plot_style()

x = np.linspace(0, 5, 50)
y = 2 * x
plt.plot(x, y)
plt.xlabel('x (input)')
plt.ylabel('y = 2x (output)')
plt.title('Linear function')
plt.show()
```

## Preview: exp, log, sin

```python
t = np.linspace(0, 2 * np.pi, 100)
print('exp(1) ≈', np.exp(1))
print('log(0.5) ≈', np.log(0.5))
plt.plot(t, np.sin(t), label='sin')
plt.plot(t, np.cos(t), label='cos')
plt.legend()
plt.title('Wavy functions (preview)')
plt.show()
```
