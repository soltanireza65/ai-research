# Reshaping Tensors Lab

```python
import torch
```

```python
x = torch.arange(12).reshape(2, 2, 3)
print('original:', x.shape)
print('flatten:', x.flatten().shape)
print('unsqueeze:', x.unsqueeze(0).shape)
print('squeeze:', torch.tensor([[[1], [2]]]).squeeze().shape)
```
