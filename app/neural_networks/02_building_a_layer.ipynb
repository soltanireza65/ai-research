# Building a Layer Lab

```python
import torch
```

```python
in_features, out_features = 4, 3
batch = 2

x = torch.randn(batch, in_features)
W = torch.randn(out_features, in_features)
b = torch.randn(out_features)

y = x @ W.T + b
print('input:', x.shape)
print('output:', y.shape)
print(y)
```
