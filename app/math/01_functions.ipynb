```python
import numpy as np
import matplotlib.pyplot as plt

from app.utils.plot import setup_plot_style

setup_plot_style()
```

## Linear: y = 2x

```python
x = np.linspace(0, 10, 100)
y = 2 * x

plt.plot(x, y, label='y = 2x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear function')
plt.legend()
plt.show()
```

## Quadratic: y = x²

```python
x = np.linspace(-3, 3, 100)
y = x**2

plt.plot(x, y, label='y = x²')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic function')
plt.legend()
plt.show()
```
