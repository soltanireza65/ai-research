## 2D vectors

```python
v1 = np.array([3, 4])
v2 = np.array([-2, 1])

fig, ax = plt.subplots()
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='b', label='v1')
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='r', label='v2')
ax.set_xlim(-4, 5)
ax.set_ylim(-1, 5)
ax.set_aspect('equal')
ax.legend()
plt.show()

print('||v1|| =', np.linalg.norm(v1))
```

## Dot product

```python
v1 = np.array([3, 4])
v2 = np.array([-2, 1])
print('v1 · v2 =', np.dot(v1, v2))
print('||v1|| =', np.linalg.norm(v1))
```
