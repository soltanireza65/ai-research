# Single Neuron Lab

```python
import torch

def relu(x):
    return torch.maximum(x, torch.tensor(0.0))
```

```python
x = torch.tensor([3.0, 4.0])
w = torch.tensor([2.0, -1.0])
b = torch.tensor(0.5)

z = (x * w).sum() + b
y = relu(z)
print('pre-activation:', z.item())
print('output:', y.item())
```
