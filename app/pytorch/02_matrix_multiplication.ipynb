## Easy: 2×3 @ 3×1

```python
A = torch.tensor([[1., 2., 3.], [4., 5., 6.]])
B = torch.tensor([[1.], [2.], [3.]])
print('A @ B =\n', A @ B)
print('shape:', (A @ B).shape)
```
