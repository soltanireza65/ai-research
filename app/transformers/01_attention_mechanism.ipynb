# Attention Mechanism Lab

```python
import torch
import torch.nn.functional as F
```

```python
Q = torch.tensor([[1., 0.], [0., 1.]])
K = torch.tensor([[1., 0.], [0., 1.]])
V = torch.tensor([[2., 3.], [4., 5.]])

d_k = Q.shape[-1]
scores = Q @ K.T / (d_k ** 0.5)
weights = F.softmax(scores, dim=-1)
out = weights @ V
print('weights:\n', weights)
print('output:\n', out)
```
